```python
# Read the data

# First the train data
x_train = pd.read_table("./UCI HAR Dataset/train/X_train.txt", delim_whitespace = True, header = None)
y_train = pd.read_table("./UCI HAR Dataset/train/y_train.txt", delim_whitespace = True, header = None)
subject_train = pd.read_table("./UCI HAR Dataset/train/subject_train.txt", delim_whitespace = True, header = None)

# Now the test data
x_test = pd.read_table("./UCI HAR Dataset/test/X_test.txt", delim_whitespace = True, header = None)
y_test = pd.read_table("./UCI HAR Dataset/test/y_test.txt", delim_whitespace = True, header = None)
subject_test = pd.read_table("./UCI HAR Dataset/test/subject_test.txt", delim_whitespace = True, header = None)

# Read the activity labels and also the list of features
features = pd.read_table("./UCI HAR Dataset/features.txt", delim_whitespace = True, header = None)
activity_labels = pd.read_table("./UCI HAR Dataset/activity_labels.txt", delim_whitespace = True, header = None)
```

```python
# Merge both data sets into a single one
x = pd.concat([x_train, x_test], axis = 0).reset_index(drop = True)
y = pd.concat([y_train, y_test], axis = 0).reset_index(drop = True)
subject = pd.concat([subject_train, subject_test], axis = 0).reset_index(drop = True)
```

```python
# Add activity labels to y data frame
y = y.merge(activity_labels, left_on = 0, right_on = 0, how = "left")

# Add columns of labels (y) and subject to dataset
x = pd.concat([x, y[1], subject], axis = 1)

# Add labels to x, y and subject
x.columns = list(features[1]) + ["Activity_Label", "Subject"]
```

```python
# Retains only the measurements for mean and standard deviation
msd_columns = x.filter(regex = "^(?!.*Freq).*mean|.*std", axis = 1)
x_msd = pd.concat([msd_columns, x.loc[:, ["Activity_Label", "Subject"]]], axis = 1)
x_msd
```

```python
# Makes the names of the variables more clear to understand
x_msd.columns = x_msd.columns.str.replace("^t","Time", regex = True)
x_msd.columns = x_msd.columns.str.replace("^f","Frequency", regex = True)
x_msd.columns = x_msd.columns.str.replace("Acc","Acceleration", regex = True)
x_msd.columns = x_msd.columns.str.replace("Gyro","Gyroscope", regex = True)
x_msd
```

```python
# Creates new data set containing the average of each variable per activity and subject
x_summary = x_msd.groupby(["Activity_Label", "Subject"]).mean()
x_summary
```

```python
# Writes the data sets to a .txt file
x_msd.to_csv("mean_and_std_py.txt", index = False)
x_summary.to_csv("summary_mean_py.txt", index=False)
```
